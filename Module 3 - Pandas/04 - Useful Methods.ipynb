.apply() Method On Multiple Columns:

```python
def simple(num):
    return num * 2
```

```python
simple(2)
```

```python
lambda num: num * 2 #Lambdas will be used to format apply methods correctly
```

```python
df['total_bill'].apply(simple)
```

```python
df['total_bill'].apply(lambda bill: bill * 2)
```

```python
df.head()
```

```python
def quality(total_bill, tip):
    if tip / total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
```

```python
quality(16.99, 1.01)
```

```python
df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis = 1) #Make sure you balance your parenthesis out!
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis = 1)
#Structure: You select the columns to use, call a lambda on the dataframe being passed in, and in the lambda, pass in your custom function, in which in turn you pass in the columns you want to use
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip']) #Faster and simpler way to do the previous cell
```

```python
df.head() #The vectorize method performs faster, despite the documentation saying that it's just for convenience
#np.vectorize transforms functions which aren't numpy-aware into numpy-aware functions
```

```python
import timeit #Testing/comparing the performance of the brute-force and vectorize methods
```

```python
# code snipper to be executed only once
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def quality(total_bill, tip):
    if tip / total_bill > 0.25:
        return 'Generous'
    else:
        return 'Other'
'''
```

```python
# code snippet whose execution time is to be measured
stmt_one = '''
df['Tip Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis = 1)'''

stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])'''
```

```python
timeit.timeit(setup = setup, stmt = stmt_one, number = 1000)
```

```python
timeit.timeit(setup = setup, stmt = stmt_two, number = 1000)
```
