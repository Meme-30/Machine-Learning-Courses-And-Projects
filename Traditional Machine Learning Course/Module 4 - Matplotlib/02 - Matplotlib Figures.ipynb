```python
import numpy as np
import matplotlib.pyplot as plt
```

Data:

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

Creating A Figure:

```python
fig = plt.figure()
```

Adding Axes:

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1]) #Origin X, Origin Y, Width, Height
```

Plotting On Axes:

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b)
```

```python
type(fig)
```

Using Multiple Axes:

```python
fig = plt.figure()

#Large Axes:
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

#Small Axes:
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(x, y)
```

Adding Titles And Labels To Axes:

```python
fig = plt.figure()

#Large Axes:
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8) #Same as functional, except you add '.set_'
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#Small Axes:
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

```python
fig = plt.figure()

#Large Axes:
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8) #Same as functional, except you add '.set_'
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#Small Axes:
axes2 = fig.add_axes([1, 1, 2, 2])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

Figure Parameters:

```python
fig = plt.figure(dpi = 200) #The dpi setting affects the clarity of the figure

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
```

```python
fig = plt.figure(figsize = (10, 3),dpi = 200) #The figsize setting stretches the figure by setting the image size in inches

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
```

Exporting A Figure:

```python
fig = plt.figure(figsize = (2, 2),dpi = 200) #The figsize setting stretches the figure by setting the image size in inches

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

fig.savefig('new_figure.png') #By default, you can't see the axes, as there's a bounding box around the figure
```

```python
fig = plt.figure(figsize = (2, 2),dpi = 200) #The figsize setting stretches the figure by setting the image size in inches

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

fig.savefig('new_figure.png', bbox_inches = 'tight')
```
