```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 * x
```

Creating Subplots:

```python
fig,axes = plt.subplots() #Defaults to 1 row and 1 column, creating a single set of axes
```

```python
axes
```

```python
fig,axes = plt.subplots()

axes.plot(x, y)
```

Creating Multiple Plots With .subplots():

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2) #Automatically aligns the subplots
```

```python
type(axes) #Axes becomes a numpy array when there's more than one subplot
```

```python
axes.shape #As such, it has a shape
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2) #Automatically aligns the subplots

axes[0].plot(x, y) #When there's 2+ subplots, you have to index the subplot you want to plot on
axes[1].plot(a, b)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2) #Automatically aligns the subplots

axes[0][0].plot(x, y) #When there's 3+ subplots spanning both multiple rows and columns, you now nested-index the subplot you want to plot on
axes[1][1].plot(a, b)
```

```python
type(axes)
```

```python
axes.shape
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 1)

for ax in axes:
    ax.plot(x, y) #More useful when the variables are lists of multiple coordinates
```

Adjusting Subplot Spacing:

```python
fig,axes = plt.subplots(nrows = 3, ncols = 1)

axes[0].plot(x, y)

plt.tight_layout() #Attempts to ensure that the axis tickers don't overlap over anything else
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

plt.tight_layout() #Does work, but sometimes may still be too close together
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

fig.subplots_adjust(wspace = 0.9) #Feel free to do more research on this function and its parameters
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

fig.subplots_adjust(wspace = 1, hspace = 1)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x, y)
axes[0][1].plot(x, y)
axes[1][0].plot(x, y)
axes[1][1].plot(a, b)

fig.subplots_adjust(wspace = 1, hspace = 0.5)
```

Adding Features To Subplots:

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel('Y LABEL 1, 0')
axes[1][0].set_xlim(2, 6)

axes[1][1].plot(a, b)
axes[1][1].set_title('TITLE 1, 1')

plt.tight_layout()
```

Figure-Level Features:

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel('Y LABEL 1, 0')
axes[1][0].set_xlim(2, 6)

axes[1][1].plot(a, b)
axes[1][1].set_title('TITLE 1, 1')

fig.suptitle('Figure Level', fontsize = 16) #Methods like these also have many parameters that you can feel free to explore

plt.tight_layout()
```

Exporting Figures With Subplots:

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel('Y LABEL 1, 0')
axes[1][0].set_xlim(2, 6)

axes[1][1].plot(a, b)
axes[1][1].set_title('TITLE 1, 1')

fig.suptitle('Figure Level', fontsize = 16)

plt.tight_layout()

fig.savefig('new_subplots.png', bbox_inches = 'tight')
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 2, figsize = (4, 10), dpi = 200)

axes[0][0].plot(x, y)

axes[0][1].plot(x, y)

axes[1][0].plot(x, y)
axes[1][0].set_ylabel('Y LABEL 1, 0')
axes[1][0].set_xlim(2, 6)

axes[1][1].plot(a, b)
axes[1][1].set_title('TITLE 1, 1')

fig.suptitle('Figure Level', fontsize = 16)

plt.tight_layout()

fig.set_figwidth(10)

fig.savefig('new_subplots.png', bbox_inches = 'tight')
```
