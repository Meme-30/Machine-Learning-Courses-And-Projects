```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

Plotting A Function:

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
plt.plot(x, y) #By default, a connecting line is drawn between the points
```

```python
plt.plot(x, y)
plt.show() #Necessary if you're running from a .py file
```

Adding Titles And Labels To Plots:

```python
plt.plot(x, y)
plt.title('String Title');
```

```python
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('String Title'); #Note: In Jupyter notebook, everything pertaining to one plot must be in the same cell
```

Adjusting Limits:

```python
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.xlim(0, 6)
plt.ylim(0, 15)
plt.title('String Title');
```
