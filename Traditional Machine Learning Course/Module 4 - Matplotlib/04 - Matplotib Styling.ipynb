```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 11, 10)
```

Plotting Multiple Functions On One Plot:

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x)

ax.plot(x, x ** 2)
```

Adding Legends:

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')

ax.plot(x, x ** 2, label = 'X vs X^2')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')

ax.plot(x, x ** 2, label = 'X vs X^2')

ax.legend(loc = 'lower left') #The loc argument moves the legend to a specific location, with different numbers corresponding to different locations
#By default, leaving it empty or 0 automatically moves it to the 'best' possible location based on what's plotted
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')

ax.plot(x, x ** 2, label = 'X vs X^2')

ax.legend(loc = 4) #4 = 'lower right'
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')

ax.plot(x, x ** 2, label = 'X vs X^2')

ax.legend(loc = (1.1, 0.5)) #Passing in a tuple sets the exact location, including outside the plot itself
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')

ax.plot(x, x ** 2, label = 'X vs X^2')

ax.legend(loc = (0.5, 0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label = 'X vs X')

ax.plot(x, x ** 2, label = 'X vs X^2')

ax.legend(loc = (-0.5, 0.5)) #Note: The location is based on the bottom-left corner of the legend
```

Changing Colors:

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = 'blue') #Basic colors have strings that you can type out to specify
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = 'lightblue')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = 'purple')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = '#650dbd') #Passing in an RGB color code as a string gives you full control of the plot's color
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = '#e6c017', label = 'x vs x') #Plot colors also show up on legends

ax.plot(x, x + 1, color = 'purple', label = 'x vs x+1')

ax.legend()
```

Line Styles:

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = '#e6c017', lw = 10) #The lw argument sets the line thickness to that many times the default thickness

ax.plot(x, x + 1, color = 'purple', linewidth = 0.5) #The full name 'linewidth' can also be typed out for the same effect
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
#Possible linestyle options: '--', '-', '-.', ':', 'steps'

ax.plot(x, x, color = '#e6c017', lw = 10, ls = '--') #The ls argument controls the linestyle

ax.plot(x, x + 1, color = 'purple', linewidth = 2, linestyle = '-.') #Typing it out also works
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

#Setting a custom line style:

lines = ax.plot(x, x + 1, color = 'purple', linewidth = 5)
lines[0].set_dashes([1, 1, 1, 2, 3, 5]) #Numbers always come in pairs for how many points you want and how much space you want after it
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x + 1, color = 'purple', linewidth = 2, ls = '--', marker = 'o')
#The marker argument sets the type of markers at the input points from your function
#By default, markers don't appear
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = 'purple', linewidth = 2, ls = '--', marker = 'o', ms = 10) #The ms argument controls the markers' size

ax.plot(x, x + 1, color = 'red', linewidth = 2, ls = '--', marker = '+', markersize = 20) #Markersize does too
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color = 'purple', linewidth = 2, ls = '--', marker = 's', ms = 15,
        markerfacecolor = 'red', markeredgewidth = '4', markeredgecolor = 'orange')
#Marker face color lets you make the markers a different color than the line
#Marker edge width adds an extra edge to the markers with a specified thickness
#Marker edge color lets you make the marker edges a different color than the middle of the marker
```
