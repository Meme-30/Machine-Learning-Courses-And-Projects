```python
df = pd.read_csv('Sales_Funnel_CRM.csv')
```

Creating A Pivot Table With .pivot():

```python
help(pd.pivot)
```

```python
licenses = df[['Company', 'Product', 'Licenses']]
```

```python
licenses
```

```python
pd.pivot(data = licenses, index = 'Company', columns = 'Product', values = 'Licenses')
```

Aggregate Methods With .pivot_table():

```python
pd.pivot_table(df, index = 'Company', aggfunc = 'sum') #Index is what you group by
#This function does work with string entries, but you should still filter to numeric columns
```

```python
numeric_cols = df.select_dtypes(include = 'number').columns
```

```python
pd.pivot_table(df, index = 'Company', aggfunc = 'sum')[numeric_cols]
```

```python
df.groupby('Company')[numeric_cols].sum() #Groupby equivalent
```

```python
pd.pivot_table(df, index = 'Company', aggfunc = 'sum', values = ['Licenses', 'Sale Price']) #Filtering select columns, which DOES NOT require numeric column filtering first
```

Multilevel Indexed Pivot Tables:

```python
pd.pivot_table(df, index = ['Account Manager', 'Contact'], values = ['Sale Price'], aggfunc = 'sum')
```

```python
pd.pivot_table(df, index = ['Account Manager', 'Contact'], values = ['Sale Price'], columns = ['Product'], aggfunc = 'sum') #2-Tiered Multilevel Indexing
```

```python
pd.pivot_table(df, index = ['Account Manager', 'Contact'], values = ['Sale Price'], columns = ['Product'], aggfunc = 'sum', fill_value = 0) #Filling null values with 0
```

```python
pd.pivot_table(df, index = ['Account Manager', 'Contact'], values = ['Sale Price'], columns = ['Product'], aggfunc = [np.sum], fill_value = 0) #Passing in numpy calls
```

```python
pd.pivot_table(df, index = ['Account Manager', 'Contact'], values = ['Sale Price'], columns = ['Product'], aggfunc = [np.sum, np.mean], fill_value = 0) #Passing in multiple numpy calls
```

```python
pd.pivot_table(df, index = ['Account Manager', 'Contact', 'Product'], values = ['Sale Price'], aggfunc = [np.sum], fill_value = 0) #3-Ties Multilevel Indexing
```

```python
pd.pivot_table(df, index = ['Account Manager', 'Contact', 'Product'], values = ['Sale Price'], aggfunc = [np.sum], fill_value = 0, margins = True) #The margins paramater adds grand totals to each aggregate function call
```
