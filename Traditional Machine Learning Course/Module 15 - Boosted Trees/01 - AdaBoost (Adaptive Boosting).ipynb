Data Preparation:

```python
X = df.drop('class', axis = 1)
```

```python
X.isnull().sum()
```

```python
X = pd.get_dummies(X, drop_first = True, dtype = int)
#All forms of decision tree algorithms require categorical data to be in dummy format
```

```python
X
```

```python
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 101)
```

Creating A Simple AdaBoost Model With A Decision Tree:

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier(n_estimators = 1)
#By default (when None/empty), the base_model parameter is a DecisionTreeClassifier with a max_depth of 1 (a stump)
#The n_estimators paramater sets the number of learners to use, where each new learner learns from the mistakes of the previous
#By default, the n_estimators parameter is 50
```

```python
model.fit(X_train, y_train)
```

Model Evaluation:

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, accuracy_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
print(classification_report(y_test, predictions))
#Shockingly, for this data set, you get pretty impressive performance considering you basically predicted based on one dummy feature
```

```python
model.feature_importances_
```

```python
model.feature_importances_.argmax() #The index of the 1, the feature that was chosen
```

Exploring More Features With AdaBoosted Stumps:

```python
len(X.columns)
```

```python
#Elbow method for finding the ideal number of stumps:
error_rates = []

for n in range(1, 96):
    model = AdaBoostClassifier(n_estimators = n)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    err = 1 - accuracy_score(y_test, preds)

    error_rates.append(err)
```

```python
plt.plot(range(1, 96), error_rates)
```

```python
model #Note that the for loop 2 cells above saves the last model of the loop
```

```python
model.feature_importances_
```

```python
feats = pd.DataFrame(index = X.columns, data = model.feature_importances_, columns = ['Importance'])
```

```python
feats
```

```python
imp_feats = feats[feats['Importance'] > 0]
```

```python
plt.figure(figsize = (14, 6), dpi = 200)
sns.barplot(data = imp_feats.sort_values('Importance'), x = imp_feats.index, y = 'Importance')
plt.xticks(rotation = 90);
```

Note: Unlike Random Forests, Boosting Methods Can Overfit, But It Takes Extreme Learner Numbers To Get There
