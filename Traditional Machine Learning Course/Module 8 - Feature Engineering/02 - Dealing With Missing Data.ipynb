Calculating NaN Counts And Ratios:

```python
df.isnull().sum()
```

```python
100 * df.isnull().sum() / len(df) #Transfroms the above into percentages
```

```python
def percent_missing(df):
    percent_nan = 100 * df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    return percent_nan
#Feel free to use this function on any dataframe
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
plt.figure(figsize = (8, 4), dpi = 200)
sns.barplot(x = percent_nan.index, y = percent_nan)
plt.xticks(rotation = 90);
```

```python
#Neighborhood: Physical locations within Ames city limits
#LotFrontage: Linear feet of street connected to property
#Neighborhood will be used as domain knowledge to help fill the null values in the Lot Frontage column
```

```python
plt.figure(figsize = (8, 12), dpi = 200)
sns.boxplot(x = 'Lot Frontage', y = 'Neighborhood', data = df, orient = 'h')
#Plan: If a Lot Frontage value is missing, but the Neighborhood is known, the average value for that neighborhood will be used
```

```python
df.groupby('Neighborhood')['Lot Frontage'].mean()
```

```python
df.groupby('Neighborhood')['Lot Frontage'].transform(lambda value: value.fillna(value.mean()))
#The .transform() function call lets you combine the groupby() call with an apply method
```

```python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda value: value.fillna(value.mean()))
```
