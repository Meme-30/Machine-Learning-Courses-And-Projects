### Creating Cohorts based on Tenure

**Let's begin by treating each unique tenure length, 1 month, 2 month, 3 month...N months as its own cohort.**

**TASK: Treating each unique tenure group as a cohort, calculate the Churn rate (percentage that had Yes Churn) per cohort. For example, the cohort that has had a tenure of 1 month should have a Churn rate of 61.99%. You should have cohorts 1-72 months with a general trend of the longer the tenure of the cohort, the less of a churn rate. This makes sense as you are less likely to stop service the longer you've had it.**

```python
tenure_cohort_series = pd.Series()

for i in range(1, 73):
    churn_rate = len(df[(df['tenure'] == i) & (df['Churn'] == 'Yes')]) / len(df[df['tenure'] == i]) * 100
    tenure_cohort_series = pd.concat([tenure_cohort_series, pd.Series(churn_rate)])

tenure_cohort_series = tenure_cohort_series.reset_index(drop = True)
tenure_cohort_series.index = range(1, len(tenure_cohort_series) + 1)
tenure_cohort_series
```

**TASK: Now that you have Churn Rate per tenure group 1-72 months, create a plot showing churn rate per months of tenure.**

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(tenure_cohort_series.index, tenure_cohort_series)
axes.set_xlabel('tenure')
axes.set_ylabel('Churn Percentage')
```

<img src='fig9.png'>

### Broader Cohort Groups
**TASK: Based on the tenure column values, create a new column called Tenure Cohort that creates 4 separate categories:**
   * '0-12 Months'
   * '12-24 Months'
   * '24-48 Months'
   * 'Over 48 Months'    

```python
tenure_cohort_df = df.copy(deep = True)
tenure_cohort_df['tenure_cohort'] = None

for index, row in tenure_cohort_df.iterrows():
    if row['tenure'] <= 12:
        tenure_cohort_df.at[index, 'tenure_cohort'] = '0-12 Months'
    elif row['tenure'] > 12 and row['tenure'] <= 24:
        tenure_cohort_df.at[index, 'tenure_cohort'] = '12-24 Months'
    elif row['tenure'] > 24 and row['tenure'] <= 48:
        tenure_cohort_df.at[index, 'tenure_cohort'] = '24-48 Months'
    else:
        tenure_cohort_df.at[index, 'tenure_cohort'] = 'Over 48 Months'

tenure_cohort_df
```

## Single Decision Tree

**TASK : Separate out the data into X features and Y label. Create dummy variables where necessary and note which features are not useful and should be dropped.**

```python
X = pd.get_dummies(df.drop(['Churn', 'customerID'], axis = 1), dtype = int, drop_first = True)
```

```python
y = df['Churn']
```

**TASK: Perform a train test split, holding out 10% of the data for testing. We'll use a random_state of 101 in the solutions notebook/video.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 101)
```

**TASK: Decision Tree Perfomance. Complete the following tasks:**
   1. Train a single decision tree model (feel free to grid search for optimal hyperparameters).
   2. Evaluate performance metrics from decision tree, including classification report and plotting a confusion matrix.
   2. Calculate feature importances from the decision tree.
   4. OPTIONAL: Plot your tree, note, the tree could be huge depending on your pruning, so it may crash your notebook if you display it with plot_tree.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
base_tree = DecisionTreeClassifier()
```

```python
param_grid = {'max_depth': [2, 3, 4, 5, 6], 'criterion': ['gini', 'entropy']}
```

```python
grid_tree = GridSearchCV(base_tree, param_grid)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
y_pred = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, y_pred))
```

```python
ConfusionMatrixDisplay.from_estimator(grid_tree, X_test, y_test)
```

```python
feat_import = grid_tree.best_estimator_.feature_importances_
```

```python
imp_feat = pd.DataFrame(index = X.columns, data = feat_import, columns = ['Importance'])
```

```python
imp_feat = imp_feat.sort_values('Importance')
```

```python
plt.figure(figsize = (12, 8), dpi = 200)
sns.barplot(data = imp_feat, x = imp_feat.index, y = 'Importance')
plt.xticks(rotation = 90);
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (12, 8), dpi = 200)
plot_tree(grid_tree.best_estimator_);
```
