```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cancer_tumor_data_features.csv')
```

```python
df.head()
```

Data Preparation:

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X
```

Performing PCA With Scikit-Learn:

```python
from sklearn.decomposition import PCA
```

```python
help(PCA)
```

```python
pca_model = PCA(n_components = 2)
#The n_components parameter dictates the number of features the target feature set has
```

```python
pca_model.fit(scaled_X) #Fitting to a PCA model calculates the eigenvalues and eigenvectors
```

```python
pca_model.transform(scaled_X) #Transwforming with a PCA model performs the projection step of the dot product
```

```python
pc_results = pca_model.fit_transform(scaled_X) #fit_transform performs both steps above at once
```

```python
plt.figure(figsize = (6, 8), dpi = 150)
plt.scatter(pc_results[:, 0], pc_results[:, 1])
```

Loading And Applying The Labels:

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dictionary = load_breast_cancer()
```

```python
plt.figure(figsize = (6, 8), dpi = 150)
plt.scatter(pc_results[:, 0], pc_results[:, 1], c = cancer_dictionary['target'])
```

```python
cancer_dictionary['target']
```

Useful Attributes:

```python
pca_model.components_
#The components_ attribute is an array that represents the directions of maximum variance in the data, sorted by explained variance
```

```python
df_comp = pd.DataFrame(pca_model.components_, index = ['PC1', 'PC2'], columns = df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize = (20, 3), dpi = 150)
sns.heatmap(df_comp, annot = True)
```

```python
pca_model.explained_variance_ratio_
#The explained_variance_ratio_ attribute gives the percent in which each principal component explains that much of the variance in the data
```

```python
np.sum(pca_model.explained_variance_ratio_)
#Taking the sum of the explained variance ratioes provides how much of the variance all of the components combined explains from the original dataset
```

Exploring Explained Variance Ratio:

```python
pca_30 = PCA(n_components = 30)
```

```python
pca_30.fit(scaled_X)
```

```python
pca_30.explained_variance_ratio_
```

```python
np.sum(pca_30.explained_variance_ratio_)
#When you don't reduce the number of components, you don't lose any variance in the data, as expected
#Intuitively, there exists a relationship between the number of principal components and the sum of explained variances
```

```python
explained_variance = []

for n in range(1, 30):

    pca = PCA(n_components = n)
    pca.fit(scaled_X)

    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1, 30), explained_variance)
plt.xlabel('Num of Components')
plt.ylabel('Variance Explained')
#As expected, the more principal components, the more variance explained
#Note that the y-axis starts around 40%
```
