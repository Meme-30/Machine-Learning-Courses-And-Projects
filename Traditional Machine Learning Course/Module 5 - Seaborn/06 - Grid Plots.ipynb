```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

Category Plots:

```python
sns.catplot(data = df, x = 'gender', y = 'math score', kind = 'box') #This function is the basis for many of the plot types you've already learned about
```

```python
sns.catplot(data = df, x = 'gender', y = 'math score', kind = 'box', row = 'lunch')
#The power of this function is that you can use the row and col parameters to plot multiple times based on categories
#This works similar to the hue parameter, except it makes a separate plot for each unique value in the category instead of stacking them nearby by color
```

```python
sns.catplot(data = df, x = 'gender', y = 'math score', kind = 'box', col = 'lunch')
```

```python
sns.catplot(data = df, x = 'gender', y = 'math score', kind = 'box', col = 'lunch', row = 'test preparation course')
```

```python
sns.catplot(data = df, x = 'gender', y = 'math score', kind = 'box', col = 'lunch', row = 'race/ethnicity')
```

```python
sns.catplot(data = df, x = 'gender', y = 'math score', kind = 'violin', col = 'lunch', row = 'race/ethnicity')
#You should only call .catplot() if you're planning to use categories to make a grid of plots
```

Pair Grids:

```python
sns.pairplot(df) #This calls .PairGrid() with all numerical columns
```

```python
sns.PairGrid(df)
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_lower(sns.kdeplot) #Note: this is very CPU intensive
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot) #Note: this is very CPU intensive
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(df, hue = 'gender') #Higher level parameters are added into the initialiation call

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(df, hue = 'gender') #Higher level parameters are added into the initialiation call

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend() #Note: You have to manually call for a legend if you want one
```

Facet Grids:

```python
sns.FacetGrid(data = df, col = 'gender', row = 'lunch')
#.catplot() stems from .FacetGrid()
```

```python
g = sns.FacetGrid(data = df, col = 'gender', row = 'lunch')
g = g.map(plt.scatter, 'math score', 'reading score', edgecolor = 'w')
g.add_legend()
```
