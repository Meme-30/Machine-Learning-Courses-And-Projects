```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

Rug Plots:

```python
sns.rugplot(x = 'salary', data = df) #This is often considered a 1-dimensional scatter plot
```

```python
sns.rugplot(x = 'salary', data = df, height = 0.5) #The height parameter specifies the amount of the y axis each tick takes up
```

```python
plt.figure(figsize = (5, 8), dpi = 200) #As always, you can edit figure-level parameters with matplotlib first
sns.rugplot(x = 'salary', data = df, height = 0.5)
```

Distribution Plots And Histograms:

```python
#DO NOT USE DISTPLOT! IT'S DEPRICATED AND WILL BE REMOVED IN A FUTURE SEABORN UPDATE!
sns.displot(data = df, x = 'salary')
```

```python
sns.displot(data = df, x = 'salary', bins = 10) #The bins parameter breaks up the x axis into a certain number of 'bins'
```

```python
sns.displot(data = df, x = 'salary', bins = 100)
```

```python
sns.set(style = 'darkgrid') #The set function is also depricated
sns.displot(data = df, x = 'salary', bins = 20)
```

```python
sns.set_theme(style = 'darkgrid') #Replace the set function with the set_theme function
sns.displot(data = df, x = 'salary', bins = 20)
```

```python
sns.set_theme(style = 'whitegrid')
sns.displot(data = df, x = 'salary', bins = 20)
```

```python
sns.set_theme(style = 'white')
sns.displot(data = df, x = 'salary', bins = 20)
```

```python
sns.set_theme(style = 'dark')
sns.displot(data = df, x = 'salary', bins = 20)
```

```python
sns.set_theme(style = 'ticks') #Note: Setting the theme applies the setting for the rest of the notebook until you change it again
sns.displot(data = df, x = 'salary', bins = 20)
```

```python
sns.displot(data = df, x = 'salary', bins = 20, color = 'red', edgecolor = 'blue') #Matplotlib arguments can also be taken in
```

```python
sns.displot(data = df, x = 'salary', bins = 20, color = 'red', edgecolor = 'blue', linewidth = 4)
```

```python
sns.displot(data = df, x = 'salary', bins = 20, color = 'red', edgecolor = 'blue', linewidth = 4, ls = '--')
```

```python
sns.histplot(data = df, x = 'salary') #Distplot is a more generalized version of histplot
```

```python
sns.displot(data = df, x = 'salary', kde = True) #The Kernel Density Estimation Plot can be automatically generated on top of the histogram via the kde argument
```

```python
sns.displot(data = df, x = 'salary', kde = True, rug = True) #Same deal with the rug plot!
```

Kernel Density Estimation (KDE) Plots:

```python
sns.kdeplot(data = df, x = 'salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns = ['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data = sample_ages, x = 'age')
```

```python
sns.displot(data = sample_ages, x = 'age', rug = True, bins = 30)
```

```python
sns.displot(data = sample_ages, x = 'age', rug = True, bins = 30, kde = True)
```

```python
sns.kdeplot(data = sample_ages, x = 'age') #By default, you get some probability outside your existing age
```

```python
sns.kdeplot(data = sample_ages, x = 'age', clip = [0, 100]) #The clip parameter terminates the kde plot at the edges of the input data
```

```python
sns.kdeplot(data = sample_ages, x = 'age', clip = [0, 100], bw_adjust = 0.01) #The bw_adjust parameter makes the kdeplot more or less sensitive to variance
```

```python
sns.kdeplot(data = sample_ages, x = 'age', clip = [0, 100], bw_adjust = 0.1)
```

```python
sns.kdeplot(data = sample_ages, x = 'age', clip = [0, 100], bw_adjust = 0.5)
```

```python
sns.kdeplot(data = sample_ages, x = 'age', clip = [0, 100], bw_adjust = 2)
```

```python
sns.kdeplot(data = sample_ages, x = 'age', clip = [0, 100], bw_adjust = 20)
```

```python
sns.kdeplot(data = sample_ages, x = 'age', clip = [0, 100], bw_adjust = 0.6, shade = True) #The shade parameter fills in the area under the kde curve, which is depricated and will be replaced by 'fill'
```

```python
sns.kdeplot(data = sample_ages, x = 'age', clip = [0, 100], bw_adjust = 0.6, fill = True)
```
