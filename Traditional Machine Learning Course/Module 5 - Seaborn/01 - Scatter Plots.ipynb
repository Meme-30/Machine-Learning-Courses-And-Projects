```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

Creating A Scatter Plot From A Pandas Dataframe:

```python
sns.scatterplot(x = 'salary', y = 'sales', data = df) #Seaborn utilizes matplotlib to create its own plots
```

Editing Seaborn Plots With Matplotlib:

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df)
```

Seaborn Scatter Plot Parameters:

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'level of education') #The hue parameter lets you color the points of a scatter plot based on another parameter
#Using the hue parameter automatically creates a legend via matplotlib
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'salary')
#Using hue with a continuous column automatically creates a gradient and major stopping points to show a color on the legend
```

https://matplotlib.org/tutorials/colors/colormaps.html

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, hue = 'level of education', palette = 'Dark2') #The palette command lets you customize to an extend what colors the points are
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, size = 'salary') #The size argument scales the points by a dataframe column's values
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, size = 'work experience')
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200) #The 's' parameter universally scales all the points by a common factor
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200, alpha = 0.2) #The alpha parameter sets the points' transparency
#0 = Invisible, 1 = Opaque
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200, style = 'level of education') #The style parameter shapes the points based on another column
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200, style = 'level of education', hue = 'level of education')
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200, style = 'level of education', hue = 'level of education', alpha = 0.3)
```

Exporting A Seaborn Figure:

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.scatterplot(x = 'salary', y = 'sales', data = df, s = 200, style = 'level of education', hue = 'level of education')

plt.savefig('my_plot.jpg') #Exactly the same as matplotlib
```
