```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

Count Plots:

```python
plt.figure(figsize = (10, 4))
sns.countplot(data = df, x = 'division')
```

```python
plt.figure(figsize = (10, 4))
sns.countplot(data = df, x = 'division')
plt.ylim(90, 260) #This isn't recommended, as the data doesn't appear to be displayed in a truthful manner
```

```python
plt.figure(figsize = (10, 4))
sns.countplot(data = df, x = 'division')
plt.ylim(0, 1000) #This also isn't recommended, as the data doesn't appear to be displayed in a truthful manner
```

```python
plt.figure(figsize = (10, 4))
sns.countplot(data = df, x = 'level of education')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize = (10, 4))
sns.countplot(data = df, x = 'level of education', hue = 'division') #The hue parameter subdivides each category by another category
```

```python
plt.figure(figsize = (10, 4))
sns.countplot(data = df, x = 'level of education', hue = 'division', palette = 'Set2') #The palette parameter lets you customize to extend what colors the bars are
```

General Bar Plots:

```python
plt.figure(figsize = (10, 4))
sns.barplot(data = df, x = 'level of education', y = 'salary', estimator = np.mean, ci = 'sd')
#The estimator parameter works like pandas' groupby method
#The ci parameter, which is depricated, sets the confidence interval, which can be selected to a certain statistic
```

```python
plt.figure(figsize = (10, 4))
sns.barplot(data = df, x = 'level of education', y = 'salary', estimator = np.mean, errorbar = 'sd') #Use the 'errorbar' parameter going forward
```

```python
plt.figure(figsize = (10, 4))
sns.barplot(data = df, x = 'level of education', y = 'salary', estimator = np.mean, errorbar = 'sd', hue = 'division')
plt.legend(bbox_to_anchor = (1.05, 1))
```
