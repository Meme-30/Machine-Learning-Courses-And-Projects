```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df #Not every dataframe is a good choice for a heat map
```

```python
df = df.set_index('Countries')
```

```python
df #This dataframe is now a good choice for a heat map
```

Heat Maps:

```python
sns.heatmap(df) #Be careful: In this example, one of the columns is in different units than the others (Life expectancy isn't a rate)
#It's recommended to have all the columns in a heatmap be of the same units
```

```python
sns.heatmap(df.drop('Life expectancy', axis = 1))
```

```python
sns.heatmap(df.drop('Life expectancy', axis = 1), linewidth = 0.5)
```

```python
plt.figure(dpi = 200)
sns.heatmap(df.drop('Life expectancy', axis = 1), linewidth = 0.5, annot = True)
# The annot parameter (annotate) adds numbers inside the boxes of the heat map
```

```python
plt.figure(dpi = 200)
sns.heatmap(df.drop('Life expectancy', axis = 1), linewidth = 0.5, annot = True, cmap = 'viridis')
#The cmap parameter is the heat map's equivalent of the palette parameter for most other plots
```

```python
plt.figure(dpi = 200)
sns.heatmap(df.drop('Life expectancy', axis = 1), linewidth = 0.5, annot = True, cmap = 'viridis', center = 40)
#The center parameter moves the 'center' of the selected color mapping to a certain percentage of the color map
#It's not recommended to use the center parameter
```

Cluster Maps:

```python
plt.figure(dpi = 200)
sns.clustermap(df.drop('Life expectancy', axis = 1), linewidth = 0.5, annot = True, cmap = 'viridis')
#Cluster maps help you determine which columns (in this case rates) are more related than others
```

```python
plt.figure(dpi = 200)
sns.clustermap(df.drop('Life expectancy', axis = 1), linewidth = 0.5, annot = True, cmap = 'viridis', col_cluster = False)
```
