```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

Box Plots:

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.boxplot(data = df, y = 'math score')
```

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.boxplot(data = df, y = 'math score', x = 'test preparation course')
```

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.boxplot(data = df, y = 'reading score', x = 'test preparation course')
```

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.boxplot(data = df, y = 'reading score', x = 'parental level of education')
```

```python
plt.figure(figsize = (10, 4), dpi = 200)
sns.boxplot(data = df, y = 'reading score', x = 'parental level of education', hue = 'test preparation course') #Notice that the hue parameter does the same thing with many seaborn plot types
plt.legend(bbox_to_anchor = (1.05, 0.5))
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.boxplot(data = df, x = 'reading score', y = 'parental level of education', hue = 'test preparation course')
#Switching the x and y parameters flips the direction of the box plots between vertical and horizontal
plt.legend(bbox_to_anchor = (1.05, 0.5))
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.boxplot(data = df, x = 'reading score', y = 'parental level of education', hue = 'test preparation course', palette = 'Set2')
plt.legend(bbox_to_anchor = (1.05, 0.5))
```

Violin Plots:

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education')
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education',  hue = 'test preparation course', palette = 'Set2')
plt.legend(bbox_to_anchor = (1.05, 0.5))
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education',  hue = 'test preparation course', palette = 'Set2', split = True)
#The split parameter makes the two halves of the violin plots represent the two separate KDEs of a binary category
plt.legend(bbox_to_anchor = (1.05, 0.5))
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education',  hue = 'test preparation course', palette = 'Set2', split = True, inner = None)
#The inner parameter toggles what is present inside the violin
plt.legend(bbox_to_anchor = (1.05, 0.5))
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education',  hue = 'test preparation course', palette = 'Set2', split = True, inner = 'quartile')
plt.legend(bbox_to_anchor = (1.05, 0.5))
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education', inner = 'quartile')
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education', inner = 'stick')
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education', bw = 0.1)
#The bw parameter, which is depricated, affects the sensitivity of the gaussian curves used to create the KDEs
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education', bw_method = 0.1)
#bw_method replaces bw
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education', bw_method = 0.5)
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education', bw_method = 0.01)
#Too sensitive!
```

```python
plt.figure(figsize = (10, 8), dpi = 200)
sns.violinplot(data = df, x = 'reading score', y = 'parental level of education', bw_method = 1)
#Not sensitive enough!
```

Swarm Plots:

```python
plt.figure(figsize = (8, 4), dpi = 200)
sns.swarmplot(data = df, x = 'math score')
```

```python
plt.figure(figsize = (8, 4), dpi = 200)
sns.swarmplot(data = df, x = 'math score', size = 2) #The size parameter increases or decreases the dot size, determining how many of the dots can actually be plotted
```

```python
plt.figure(figsize = (8, 4), dpi = 200)
sns.swarmplot(data = df, x = 'math score', y = 'gender', size = 2)
```

```python
plt.figure(figsize = (8, 4), dpi = 200)
sns.swarmplot(data = df, x = 'math score', y = 'gender', size = 2, hue = 'test preparation course')
plt.legend(bbox_to_anchor = (1.35, 0.5))
```

```python
plt.figure(figsize = (8, 4), dpi = 200)
sns.swarmplot(data = df, x = 'math score', y = 'gender', size = 2, hue = 'test preparation course', dodge = True)
#The dodge parameter splits the swarm plots by the hue category
plt.legend(bbox_to_anchor = (1.35, 0.5))
```

Boxen Plots:

```python
sns.boxenplot(x = 'math score', y = 'test preparation course', data = df)
```

```python
sns.boxenplot(x = 'math score', y = 'test preparation course', data = df, hue = 'gender')
```
