```python
import random

class Deck:
    suits = ['hearts', 'diamonds', 'clubs', 'spades']
    values = ['Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King']

    def __init__(self):
        self.cards = []
        for suit in Deck.suits:
            for value in Deck.values:
                self.cards.append(Card(suit, value))
    
    def shuffle(self):
        random.shuffle(self.cards)
    
    def deal(self):
        if len(self.cards) == 0:
            return None
        else:
            card_to_deal = self.cards[-1]
            del self.cards[-1]
            return card_to_deal
    
    def count_remaining(self):
        return len(self.cards)
    
    def get_remaining(self):
        remaining_cards = []
        for card in self.cards:
            remaining_cards.append(card.present())
        return remaining_cards

class Card:
    def __init__(self, suit, value):
        self.suit = suit
        self.value = value
    
    def present(self):
        return self.value + ' of ' + self.suit
```

```python
deck_1 = Deck()
deck_1.shuffle()
print('Dealt card =', deck_1.deal().present())
print(deck_1.count_remaining(), 'cards left')
remaining_cards = deck_1.get_remaining()
for card in remaining_cards:
    print(card)
```
